```python
import pandas as pd
#Asks how many classses you have and what your grade for each is then tells you the letter grade and GPA


#Welcome
print("Welcome, this is a GPA calculator! Please answer the following questions: ")

#Number of classes
n_class = input("How many classes are you taking: ")

one_done = 1


#Names of classes //1

records = {} #Makes an empty list

while one_done <= (int(n_class)): #Limited loop based on n_done
  class_name = input(f"{one_done}. What is the class name? \n   ").title()
  grade = int(input(f"   What was your grade for {class_name}? \n   "))
  records[class_name] = grade #Each grade is assigned to name
  one_done += 1

#Assigns letter to each grade //2
def letter_grade(grade):
  if grade >= 93:
      return 'A', 4.0
  elif grade >= 90:
      return 'A-', 3.7
  elif grade >= 87:
      return 'B+', 3.3
  elif grade >= 83:
      return 'B', 3.0
  elif grade >= 80:
      return 'B-', 2.7
  elif grade >= 77:
      return 'C+', 2.3
  elif grade >= 73:
      return 'C',  2.0
  elif grade >= 70:
      return 'C-', 1.7
  else:
      return 'F', 1.0

#Prints class, grade, and letter_grade //4
print("\nClass   |  Grade  | Letter Grade | GPA points")
print("-----------------------------------------------")


#Organize
four_done = 1
gpa_values = []
while four_done <= (int(n_class)):
  for class_name, grade in records.items():
    print(f"{class_name}    |   {grade}  |   {letter_grade(grade)[0]}   |   {letter_grade(grade)[1]}")
    four_done += 1


#Gpa calculation //5

gpa_final = (sum(gpa_values)) / (int(n_class))

print(f"\nYour Total GPA is {gpa_final}")
```
